# CNN - Model Training

```python
x_train=training[:,1:]/255
y_train=training[:,0]

x_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_validate,y_train,y_validate=train_test_split(x_train,y_train,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# * unpack the tuple
x_train = x_train.reshape(x_train.shape[0], *(28, 28, 1))
x_test = x_test.reshape(x_test.shape[0], *(28, 28, 1))
x_validate = x_validate.reshape(x_validate.shape[0], *(28, 28, 1))
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_validate.shape
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from keras.layers import Flatten,Dense,Dropout
from keras.callbacks import TensorBoard
from keras.optimizers import Adam
```

```python
cnn_model=Sequential()
```

```python
cnn_model.add(Conv2D(32,3,3,input_shape=(28,28,1), activation='relu'))
```

```python
cnn_model.add(MaxPooling2D(pool_size = (2, 2)))
```

```python
cnn_model.add(Dropout(0.25))
```

```python
cnn_model.add(Flatten())
```

```python
cnn_model.add(Dense(32,activation='relu' ) ) # reduced to 32 feature, hidden layer
```

```python
cnn_model.add(Dense(32,activation='sigmoid'))
```

```python
cnn_model.summary()
```

```python
#After defining our model and staking layrs, we need to config model. we this in compilation phase as below. 
#define loss function,optimisers and metrics for prediction.
#cnn_model.compile(
 #   loss='categorial_crossentropy',
  #  optimizer=Adam(learning_rate=0.001),
   # metrics=['accuracy'])
cnn_model.compile(optimizer='adam',
              loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
cnn_model.fit(x_train, y_train, epochs=50,validation_data=(x_validate, y_validate))
```
