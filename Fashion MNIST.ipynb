```python
#To print class names, I'm creating dictionary
class_names={0:'T-shirt/top',1:'Trouser',2:"Pullover",3:"Dress",4:"Coat",5:"Sandal",6:"Shirt",7:"Sneaker",8:"Bag",9:"Ankle boot",}

# see random image
import random
i=random.randint(0,59999)

plt.imshow(training[i,1:].reshape(28,28))

#retrive label from dictionay
f_class=class_names[int(training[i,0])]
print('label=',f_class)
# 10 classes decoding is as follows:
# 0 => T-shirt/top
# 1 => Trouser
# 2 => Pullover
# 3 => Dress
# 4 => Coat
# 5 => Sandal
# 6 => Shirt
# 7 => Sneaker
# 8 => Bag
# 9 => Ankle boot
```

```python
#View More images in grid format

#defining grid dimensions
W_grid = 12
L_grid = 4

# subplot returns the figure object and axes object
# we can use the axes object to plot specific figures at various locations
fig, axes = plt.subplots(L_grid, W_grid, figsize = (15,5))

axes= axes.ravel() # flaten the 15 x 6 matrix into 90 array

for i in np.arange(0, W_grid * L_grid):
    #selecting random number to display random example
    index = np.random.randint(0, 59999) 

    # read and display an image with the selected index/class 
    axes[i].imshow( training[index,1:].reshape((28,28)) )
    f_class=class_names[int(training[index,0])]
   
    axes[i].set_title(f_class, fontsize = 10)
    axes[i].axis('off')
plt.subplots_adjust(hspace=0.4)
```
